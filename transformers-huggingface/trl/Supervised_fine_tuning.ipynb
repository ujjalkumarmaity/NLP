**Relation to next token prediction**. Interestingly, SFT is not much different from language model pretraining—both pretraining and SFT use next token prediction as their underlying training objective! The main difference arises in the data that is used. During pretraining, we use a massive corpus of raw textual data to train the model. In contrast, SFT uses a supervised dataset of high-quality LLM outputs. During each training iteration, we sample several examples, then fine-tune the model on this data using a next token prediction objective. Typically, the next token prediction objective is only applied to the portion of each example that corresponds to the LLM’s output

## SFTTrainer

```python
from trl import SFTTrainer
from datasets import load_dataset
from transformers import AutoModelForCausalLM
```

```python
dataset = load_dataset("imdb", split="train")
dataset[0]
```

## fine-tune using SFTTrainer

```python
model = AutoModelForCausalLM.from_pretrained("facebook/opt-350m")
```

```python
from trl import SFTConfig, SFTTrainer

cfg = SFTConfig(
    output_dir="./content/",
    max_seq_length = 512,
    per_device_train_batch_size = 16,
    per_device_eval_batch_size = 16,
    gradient_accumulation_steps = 4,
    learning_rate = 0.00005,
    logging_steps=100,
    num_train_epochs=1,
    gradient_checkpointing=True,
    fp16 = True,
    warmup_steps=100
)
```

```python
trainer = SFTTrainer(model,
                     train_dataset = dataset,
                     dataset_text_field = "text",
                     max_seq_length = 512,
                    args=cfg
                     )
```

```python
trainer.train()
```

```python
trainer.model.save_pretrained("finetune-opt-350m")
```

```python
del model,trainer
```

## Train on completions only

we use `DataCollatorForCompletionOnlyLM`, specifically to completions task.

When packing=False, each sequence is processed independently, without being concatenated with other sequences

```python
from datasets import load_dataset
dataset = load_dataset("timdettmers/openassistant-guanaco", split="train")
dataset
```

```python
dataset[1]
```

```python
from transformers import AutoModelForCausalLM,AutoTokenizer
model = AutoModelForCausalLM.from_pretrained("facebook/opt-350m")
tokenizer = AutoTokenizer.from_pretrained("facebook/opt-350m")
```

```python
from trl import DataCollatorForCompletionOnlyLM
instruct_template = "### Human:"
response_template = "### Assistant:"
data_collator = DataCollatorForCompletionOnlyLM(response_template=response_template,
                                                instruction_template = instruct_template,
                                                mlm=False,
                                                tokenizer = tokenizer)
```

```python
text1 = """### Human: Hello all this should be masked.### Assistant: I should not be masked."""
text2 = """### Human: Hello all this should be masked.### Assistant: I should not be masked.### Human: All this should be masked too.### Assistant: I should not be masked too."""
encoded_text1 = tokenizer(text1)
encoded_text2 = tokenizer(text2)
examples = [encoded_text1, encoded_text2]
batch = data_collator(examples)
batch
```

So for `text1` example, input is - `### Human: Hello all this should be masked.` and output is `I should not be masked.` rest token label is `-100` means ignore these token while predict

```python
tokenizer.decode(batch['input_ids'][1])
```

```python
labels = batch["labels"]
input_ids = batch["input_ids"]

non_masked_tokens1 = input_ids[0][labels[0] != -100]
result_text1 = tokenizer.decode(non_masked_tokens1)
result_text1
```

```python
non_masked_tokens1 = input_ids[1][labels[1] != -100]
result_text1 = tokenizer.decode(non_masked_tokens1)
result_text1
```

```python
input_ids[0]
```

```python
trainer = SFTTrainer(
    model,
    args=SFTConfig(output_dir="/"),
    train_dataset=dataset,
    data_collator=data_collator,
)

trainer.train()
```
